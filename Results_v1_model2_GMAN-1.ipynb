# Learning curve

```python

losses_G = torch.load('/home/huawei/mas_implementation/acGAN_mnist/saved'+saved_v+'/losses_G_v1.pt')
losses_D = torch.load('/home/huawei/mas_implementation/acGAN_mnist/saved'+saved_v+'/losses_D_v1.pt')

plt.figure(figsize=(10,5))
plt.title("Generator and Discriminator Loss During Training")
plt.plot(losses_G,label="G")
for i in range(0,len(losses_D)):
    plt.plot(losses_D[i],label="D"+str(i+1))
plt.xlabel("iterations")
plt.ylabel("Loss")
plt.legend()
plt.savefig('/home/huawei/mas_implementation/acGAN_mnist/saved'+saved_v+'/losses')
```
